```python
def cycles_workflow_solver(max_node_num):
    data = pd.DataFrame(columns=['n','V','E', 'ET'])
    generator = AppGenerator()
    for i in range(1, max_node_num+1):
        G=generator.cycles_workflow_generator(i)
        V=len(G.nodes)
        E=len(G.edges)
        App = ServerlessAppWorkflow(G=G.copy(), delayType='None')
        start_timestamp = time.time()
        App.update_NE()
        Appcost = App.get_avg_cost()
        App.get_simple_dag()
        App.get_avg_rt()
        end_timestamp = time.time()
        ET=(end_timestamp-start_timestamp)*1000
        data = data.append(pd.Series({'n':i,'V':V,'E':E,'ET':ET}),ignore_index=True)
        data.to_csv("results.csv", index=False)
```

```python
cycles_workflow_solver(128)
```
